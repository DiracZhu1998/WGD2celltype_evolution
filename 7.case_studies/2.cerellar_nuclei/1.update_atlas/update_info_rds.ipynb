```r
load('../GSM4873767_chicken_data.RData')
chicken <- ex
DefaultAssay(chicken) <- 'RNA'
chicken@meta.data$cell_type <- sub("\\d+$", "", chicken@meta.data$merged.res.3.m.renamed)
chicken@meta.data$classes <- sub(".*\\.()", "", chicken@meta.data$cell_type)
saveRDS(chicken, file = "GSM4873767_chicken_data.rds")
```

```r
getUMAP <- function(object){
    object <- FindVariableFeatures(object, selection.method = "vst", nfeatures = 2000)
    object <- ScaleData(object)
    object <- RunPCA(object, npcs = 20)
    object <- FindNeighbors(object, dims = 1:10)
    object <- RunUMAP(object, dims=1:10)
    return(object)
}
```

```r
Mmus <- readRDS('GSM4873765_mouse_data.coding_genes.rds')
Hsap <- readRDS('GSM4873766_human_data.coding_genes.rds')
Ggal <- readRDS('GSM4873767_chicken_data.coding_genes.rds')

Mmus <- getUMAP(Mmus)
Hsap <- getUMAP(Hsap)
Ggal <- getUMAP(Ggal)
```

```r
saveRDS(Mmus,'GSM4873765_mouse_data.coding_genes.rds')
saveRDS(Hsap,'GSM4873766_human_data.coding_genes.rds')
saveRDS(Ggal,'GSM4873767_chicken_data.coding_genes.rds')
```
